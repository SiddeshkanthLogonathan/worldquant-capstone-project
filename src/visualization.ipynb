```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
performance_dataset_djia = pd.read_csv("data/processed/performances_djia.csv", index_col=0)
performance_dataset_djia.index = pd.to_datetime(performance_dataset_djia.index)
```

```python
performance_dataset_djia.tail()
```

```python
# plt.figure(figsize=(10,5))
plt.plot(performance_dataset_djia.SP100, label='DJI index')
plt.plot(performance_dataset_djia.MV, label='Mu Var')
plt.plot(performance_dataset_djia.DRL_EIIE, label='EIIE')
plt.plot(performance_dataset_djia.DRL_HYBRID_TRANSFORMER, label='Hybrid-Transformer')
plt.grid()
plt.legend()
plt.title("Deep Reinforcement Learning MPO (Portfolio Performances) - DJIA")
plt.xlabel("Date")
plt.ylabel("Rate of Return (%)")
plt.show()
```
