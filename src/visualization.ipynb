```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
performance_dataset = pd.read_csv("data/processed/performances_sp100.csv", index_col=0)
performance_dataset.index = pd.to_datetime(performance_dataset.index)
```

```python
performance_dataset.tail()
```

```python
plt.figure(figsize=(10,5))
plt.plot(performance_dataset.SP100, label='SP100 index')
plt.plot(performance_dataset.MV, label='Mu Var')
plt.plot(performance_dataset.DRL_EIIE, label='EIIE')
# plt.plot(performance_dataset.DRL_TRANSFORMER, label='Transformer')
plt.plot(performance_dataset.DRL_HYBRID_TRANSFORMER, label='Hybrid-Transformer')
plt.grid()
plt.legend()
plt.title("Deep Reinforcement Learning MPO (Portfolio Performances) - SP100")
plt.xlabel("Date")
plt.ylabel("Return (1x)")
plt.show()
```
